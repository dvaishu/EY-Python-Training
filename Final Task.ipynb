```python
inventory = {
    1: {'name': 'Oranges', 'MRP': 2.00, 'quantity': 100, 'size': '1 dozen'},
    2: {'name': 'Grapes', 'MRP': 2.75, 'quantity': 150, 'size': '500g'},
    3: {'name': 'Cereal', 'MRP': 4.50, 'quantity': 60, 'size': '500g'},
    4: {'name': 'Sugar', 'MRP': 1.80, 'quantity': 200, 'size': '1 kg'},
    5: {'name': 'Coffee', 'MRP': 5.50, 'quantity': 40, 'size': '250g'},
    6: {'name': 'Rice', 'MRP': 3.25, 'quantity': 120, 'size': '2 kg'},
    7: {'name': 'Pasta', 'MRP': 2.00, 'quantity': 80, 'size': '500g'},
    8: {'name': 'Tomatoes', 'MRP': 1.00, 'quantity': 100, 'size': '1 kg'},
    9: {'name': 'Potatoes', 'MRP': 1.25, 'quantity': 120, 'size': '1 kg'},
    10: {'name': 'Chicken', 'MRP': 6.00, 'quantity': 50, 'size': '1 kg'},
    11: {'name': 'Cheese', 'MRP': 4.75, 'quantity': 60, 'size': '250g'},
    12: {'name': 'Butter', 'MRP': 2.25, 'quantity': 70, 'size': '200g'},
    13: {'name': 'Onions', 'MRP': 0.75, 'quantity': 150, 'size': '1 kg'},
    14: {'name': 'Apples', 'MRP': 2.50, 'quantity': 100, 'size': '1 kg'},
    15: {'name': 'Bananas', 'MRP': 1.50, 'quantity': 150, 'size': '1 dozen'},
    16: {'name': 'Bread', 'MRP': 3.00, 'quantity': 50, 'size': '1 loaf'},
    17: {'name': 'Salmon', 'MRP': 7.50, 'quantity': 40, 'size': '250g'},
    18: {'name': 'Tea', 'MRP': 3.75, 'quantity': 100, 'size': '100g'},
    19: {'name': 'Toothpaste', 'MRP': 2.25, 'quantity': 60, 'size': '150g'},
}

# Initialize variables for order tracking
order_id = 1
orders = {}

def display_inventory():
    print("Inventory:")
    print("ID  Name       MRP    Quantity  Size")
    for item_id, details in inventory.items():
        print(f"{item_id:<4} {details['name']:<10} {details['MRP']:.2f} {details['quantity']:<9} {details['size']}")

def purchase_items():
    global order_id
    order = {}
    
    while True:
        display_inventory()
        item_id = int(input("Enter the ID of the item you want to purchase (0 to exit): "))
        
        if item_id == 0:
            break

        if item_id not in inventory:
            print("Invalid item ID. Please enter a valid ID.")
        else:
            quantity = int(input(f"Enter the quantity of {inventory[item_id]['name']} you want to purchase: "))
            if quantity <= 0 or quantity > inventory[item_id]['quantity']:
                print("Invalid quantity or insufficient stock. Please enter a valid quantity.")
            else:
                order[inventory[item_id]['name']] = {'quantity': quantity, 'MRP': inventory[item_id]['MRP']}
                inventory[item_id]['quantity'] -= quantity
                print("Item added to the cart.")

    if order:
        print("\nOrder Summary:")
        total_amount = 0
        for item, details in order.items():
            item_total = details['MRP'] * details['quantity']
            total_amount += item_total
            print(f"{item}: {details['quantity']} x {details['MRP']} = {item_total:.2f}")
        print(f"Total Amount: {total_amount:.2f}")

        # Save the order to a file
        with open(f"order_{order_id}.txt", "w") as order_file:
            order_file.write("Order Summary:\n")
            for item, details in order.items():
                order_file.write(f"{item}: {details['quantity']} x {details['MRP']} = {details['quantity'] * details['MRP']:.2f}\n")
            order_file.write(f"Total Amount: {total_amount:.2f}")
        
        order_id += 1

while True:
    print("\nWelcome to Grocery Store")
    print("1. Purchase Items")
    print("2. Exit")
    choice = int(input("Enter your choice: "))
    
    if choice == 1:
        purchase_items()
    elif choice == 2:
        break
    else:
        print("Invalid choice. Please enter a valid option.")
```
